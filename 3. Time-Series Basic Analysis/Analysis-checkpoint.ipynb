### ADF without differncing

```python
### For 1st Sub Period

from numpy import random
z = random.normal(size=(5,5))
Normality_df=pd.DataFrame(z)
Normality_df["Variable"]=""
Normality_df["Test Statistics"]=""
Normality_df["p-value"]=""
Normality_df["No. of lags"]=""
Normality_df["Number of observations"]=""
Normality_df["critical value (1%)"]=""
Normality_df["critical value (5%)"]=""
Normality_df["critical value (10%)"]=""

Variable=[]
Test_Statistics=[]
p_value=[]
No_of_lags_used=[]
Number_of_observations=[]
critical_value_1=[]
critical_value_5 =[]
critical_value_10=[]
H=[]

for i in ["SENSEX 50","IIP","Inflation Y-o-Y","EXI", "Lending Rate" ]:
   
    Variable.append(i)
    
    y=Data_Subsection_1[i]
        
    adft = statsmodels.tsa.stattools.adfuller(y,autolag='AIC')
    # output for dft will give us without defining what the values are.
    #hence we manually write what values does it explains using a for loop
    output = pd.Series(adft[0:4],index=['Test Statistics','p-value','No. of lags','Number of observations'])
    for key,values in adft[4].items():
        output['critical value (%s)'%key] =  values
    #output
    
    Test_Statistics.append(output[0])
    p_value.append(output[1])
    No_of_lags_used.append(output[2])
    Number_of_observations.append(output[3])
    critical_value_1.append(output[4])
    critical_value_5.append(output[5])
    critical_value_10.append(output[6])
    if output[1] > 0.05:
        H.append("Null not rejected")
    else:
        H.append("Null rejected")
    
Normality_df["Variable"]=Variable
Normality_df["Test Statistics"]=Test_Statistics
Normality_df["p-value"]=p_value
Normality_df["No. of lags"]=No_of_lags_used
Normality_df["Number of observations"]=Number_of_observations
Normality_df["critical value (1%)"]=critical_value_1
Normality_df["critical value (5%)"]=critical_value_5
Normality_df["critical value (10%)"]=critical_value_10
Normality_df["Hypothesis"]=H

Normality_df.drop([0,1,2,3,4],axis=1, inplace=True)
Normality_df
```
